```python
import glob

def search_files(directory, filename):
    search_pattern = f"{directory}/**/{filename}"
    file_list = glob.glob(search_pattern, recursive=True)
    return file_list

directory_to_search = "/home/sjoshi/anaconda3/envs/clip/.guild/runs/"
filename_to_find = "SpuCoSun.csv"
```

```python
import pandas as pd 

results = []
for f in search_files(directory_to_search, filename_to_find):
    results.append(pd.read_csv(f))
```

```python
combined_results = pd.DataFrame(results[0].columns)
for df in results:
    combined_results = combined_results.append(df, ignore_index=True)
```

```python
combined_results = combined_results.drop(0, axis=1).dropna()
combined_results = combined_results.reset_index()
```
